```python
import os
from datetime import datetime, timedelta

# Config
chunk_size = 400000
scrub_file = "scrub.txt"

# Start date
current_date = datetime.strptime("29102025", "%d%m%Y")

# Read all lines from scrub.txt
with open(scrub_file, "r") as f:
    lines = f.readlines()

# Split into chunks of 400000
index = 0
while len(lines) - index >= chunk_size:
    chunk = lines[index:index + chunk_size]
    filename = current_date.strftime("%d%m%Y") + ".txt"

    # Write chunk to new file
    with open(filename, "w") as out:
        out.writelines(chunk)
    print(f"Created {filename} with {len(chunk)} lines")

    # Move to next chunk
    index += chunk_size    
    current_date += timedelta(days=1)

# Keep remaining lines (less than 400000) back in scrub.txt
remaining = lines[index:]
with open(scrub_file, "w") as f:
    f.writelines(remaining)

print(f"Remaining {len(remaining)} lines kept in {scrub_file}")
```
